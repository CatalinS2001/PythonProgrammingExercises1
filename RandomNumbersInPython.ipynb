## Random numbers in python

```python
#1
import random
num = random.randint(0, 100)
print("your random number is",num)

#2
from random import randint
num = randint(0, 100)
print("your number is ", num)
```

```python
#Get a random number in the five times table
num = random.randrange(0, 50,5)

print(num)
```

```python
#Randrange does not include the limts
num = random.randrange(1, 10)

print(num)
```

```python
#Randint is inclusive

x = random.randint(1, 10)

print(x)
```

```python
#This shows how the random generator works
num = random.getstate()

print(num)
```

```python
#This creates a dice
diceRoll = random.randint(1,6)
print(diceRoll)
```
